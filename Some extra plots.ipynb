```r
library(ggplot2)
library(dplyr)

myday = read.csv("Bike-Sharing-Dataset//day.csv", header = TRUE)
```

```r
ggplot(myday, aes(x = cnt, y = hum)) + 
  geom_point(alpha = 0.7) + 
  geom_smooth(method = "lm", se = FALSE) + 
  labs(title = "Scatterplot of Total Rental Bikes and Humidity", 
       x = "Total Rental Bikes (count)", y = "Humidity") +
theme(axis.title.x = element_text(size = 11), 
    axis.title.y = element_text(size = 11), 
    plot.title = element_text(size = 15, hjust = 0.5),
    legend.position = "top", legend.justification = "right",
    legend.text = element_text(size = 7),
    axis.text.x = element_text(size = 7),
    axis.text.y = element_text(size = 7))


ggplot(myday, aes(x = cnt, y = temp)) + 
  geom_point(alpha = 0.7) + 
  geom_smooth(method = "lm", se = FALSE) + 
  labs(title = "Scatterplot of Total Rental Bikes and Temperature", 
       x = "Total Rental Bikes (count)", y = "Temperature (C)") +
theme(axis.title.x = element_text(size = 11), 
    axis.title.y = element_text(size = 11), 
    plot.title = element_text(size = 15, hjust = 0.5),
    legend.position = "top", legend.justification = "right",
    legend.text = element_text(size = 7),
    axis.text.x = element_text(size = 7),
    axis.text.y = element_text(size = 7))


ggplot(myday, aes(x = cnt, y = windspeed)) + 
  geom_point(alpha = 0.7) + 
  geom_smooth(method = "lm", se = FALSE) + 
  labs(title = "Scatterplot of Total Rental Bikes and Windspeed", 
       x = "Total Rental Bikes (count)", y = "Windspeed") +
theme(axis.title.x = element_text(size = 11), 
    axis.title.y = element_text(size = 11), 
    plot.title = element_text(size = 15, hjust = 0.5),
    legend.position = "top", legend.justification = "right",
    legend.text = element_text(size = 7),
    axis.text.x = element_text(size = 7),
    axis.text.y = element_text(size = 7))
```

```r
library(ggplot2)

# Define a function for creating scatterplots
create_scatterplot <- function(data, x_var, y_var, y_lab) {
  ggplot(data, aes(x = !!sym(x_var), y = !!sym(y_var))) + 
    geom_point(alpha = 0.7) + 
    geom_smooth(method = "lm", se = FALSE) + 
    labs(title = paste("Scatterplot of Total Rental Bikes and", y_lab),
         x = "Total Rental Bikes (count)", y = y_lab) +
    theme(axis.title.x = element_text(size = 11), 
          axis.title.y = element_text(size = 11), 
          plot.title = element_text(size = 15, hjust = 0.5),
          legend.position = "top", legend.justification = "right",
          legend.text = element_text(size = 7),
          axis.text.x = element_text(size = 7),
          axis.text.y = element_text(size = 7))
}

# Use the function to create scatterplots for the three variables
create_scatterplot(myday, "cnt", "hum", "Humidity")
create_scatterplot(myday, "cnt", "temp", "Temperature (C)")
create_scatterplot(myday, "cnt", "windspeed", "Windspeed")
```

```r
library(ggplot2)
library(cowplot)

# Define a function for creating scatterplots
create_scatterplot <- function(data, x_var, y_var, y_lab) {
  ggplot(data, aes(x = !!sym(x_var), y = !!sym(y_var))) + 
    geom_point(alpha = 0.7) + 
    geom_smooth(method = "lm", se = FALSE) + 
    labs(title = paste("Total Rental Bikes vs", y_lab),
         x = "Total Rental Bikes", y = y_lab) +
    theme(axis.title.x = element_text(size = 11), 
          axis.title.y = element_text(size = 11), 
          plot.title = element_text(size = 12, hjust = 0.5),
          legend.position = "top", legend.justification = "right",
          legend.text = element_text(size = 7),
          axis.text.x = element_text(size = 7),
          axis.text.y = element_text(size = 7))
}

# Use the function to create scatterplots for the three variables
p1 <- create_scatterplot(myday, "cnt", "hum", "Humidity")
p2 <- create_scatterplot(myday, "cnt", "temp", "Temperature (C)")
p3 <- create_scatterplot(myday, "cnt", "windspeed", "Windspeed")

# Arrange the plots in a single row with custom widths
plot_row <- plot_grid(p1, p2, p3, ncol = 3, widths = c(1.5, 1.5, 2))

# Display the plots
plot_row
```
